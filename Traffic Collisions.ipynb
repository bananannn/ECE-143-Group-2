## Traffic collision - preprocessing data

```python
import pandas as pd
traffic = pd.read_csv('Traffic_Collision_Data_from_2010_to_Present.csv')

```

```python
traffic['Date Occurred'] = pd.to_datetime(traffic['Date Occurred'])
traffic['year'], traffic['month'] = traffic['Date Occurred'].dt.year, traffic['Date Occurred'].dt.month
```

```python
traffic['lon'] = traffic['Location'].apply( lambda x: float(x.split(',')[0][1:]) )
traffic['lat'] = traffic['Location'].apply( lambda x: float(x.split(',')[1][:-1]) )
```

```python
traffic = traffic[(traffic['year'] == 2017) & (traffic['lat'] != 0) & (traffic['lon'] != 0)]

traffic
```

### Uber movement id

```python
import pandas as pd

map_LA = pd.read_json('LA_map.json', orient='records')
map_LA_header = pd.DataFrame(map_LA.features.values.tolist())
map_LA_coor = dict(pd.DataFrame(map_LA_header.geometry.tolist())['coordinates']) # the index off by 1

rows = []
for i in range(len(map_LA_coor)):
    row = []
    row.append(i + 1)
    
    if len(map_LA_coor[i][0][0]) == 2:
        row.append(round(map_LA_coor[i][0][0][1], 4))
        row.append(round(map_LA_coor[i][0][0][0], 4))   
        
    else:   
        row.append(round(map_LA_coor[i][0][0][0][1], 4))
        row.append(round(map_LA_coor[i][0][0][0][0], 4))
    rows.append(row)
record = pd.DataFrame(rows, columns = ['movement_id','long','lat'])

record

```

```python
example = traffic.groupby('Area Name')['Area Name'].count()

example
```

```python
area_name = list(example.keys())
```

```python
import pandas as pd

area_name_rows = []
for i in area_name:
    area_row = []
    area_row.append(i)
    lats = list(traffic[traffic['Area Name'] == i]['lat'])
    lons = list(traffic[traffic['Area Name'] == i]['lon'])
    area_row.append(min(lats))
    area_row.append(max(lats))

    area_row.append(min(lons))
    area_row.append(max(lons))
    
    area_name_rows.append(area_row)
    
area_name_table = pd.DataFrame(area_name_rows, columns = ['Area_Name','min_lat','max_lat', 'min_lon', 'max_lon'])
```

```python
area_name_table
```

```python
new = area_name_table.drop([0,7,13,15,17,20],axis=0)

# 1 & 9
new.at[1,'min_lat'] = min(new.loc[[1,9],'min_lat'])
new.at[1,'max_lat'] = max(new.loc[[1,9],'max_lat'])
new.at[1,'min_lon'] = min(new.loc[[1,9],'min_lon'])
new.at[1,'max_lon'] = max(new.loc[[1,9],'max_lon'])
new = new.drop(9,axis=0)

# 6 & 8
new.at[6,'min_lat'] = min(new.loc[[6,8],'min_lat'])
new.at[6,'max_lat'] = max(new.loc[[6,8],'max_lat'])
new.at[6,'min_lon'] = min(new.loc[[6,8],'min_lon'])
new.at[6,'max_lon'] = max(new.loc[[6,8],'max_lon'])
new = new.drop(8,axis=0)

#10
#12
#18
#19

new.reset_index(inplace=True)
new
```

```python
from bokeh.plotting import figure, show, output_file
from bokeh.palettes import cividis

output_file('draft_2.html')

p = figure(plot_width=400, plot_height=400)
#num_square = 21
#p = figure(title = '')
# p.quad(top=list(area_name_table['max_lon']), bottom=list(area_name_table['min_lon']), 
#        left=list(area_name_table['min_lat']), right=list(area_name_table['max_lat']), color=cividis(21))

x_axis_s = []
y_axis_s = []

max_num = new.shape[0]
for i in range(max_num):
    x_axis = []
    x_axis.append(new['min_lat'][i])
    x_axis.append(new['max_lat'][i])
    x_axis.append(new['max_lat'][i])
    x_axis.append(new['min_lat'][i])
    x_axis.append(new['min_lat'][i])
    x_axis_s.append(x_axis)

    y_axis = []
    y_axis.append(new['min_lon'][i])
    y_axis.append(new['min_lon'][i])
    y_axis.append(new['max_lon'][i])
    y_axis.append(new['max_lon'][i])
    y_axis.append(new['min_lon'][i])
    y_axis_s.append(y_axis)


p.multi_line(x_axis_s, y_axis_s,
            color=cividis(max_num), line_width=4)


show(p)

#x_axis_s
```
